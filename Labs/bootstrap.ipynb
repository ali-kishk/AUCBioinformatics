Bootstrap Resampling
===
*Demonstration of hypothesis testing using bootstrap resampling in R*
---
(*Interactive ShinyApp version is available [here](http://bit.ly/bootstrap_resampling)*)

Hypertension Dataset
---
*Adopted from an example by [Śaunak Sen](http://www.epibiostat.ucsf.edu/biostat/sen/)*


Systolic blood pressure was measured in progeny from a backcross between two mouse strains. 50 (randomly chosen) mice were genotyped at the [D4Mit214 marker](http://www.informatics.jax.org/marker/MGI:92846). We want to detect association between the [D4Mit214](http://www.informatics.jax.org/marker/MGI:92846) marker genotype and blood pressure. The values show the systolic blood pressure (in mm of Hg) by the marker genotype, heterozygous (*BA*) and homozygous (*BB*).

\begin{equation}
H_O: \mu_{BB} = \mu_{BA} \\
H_A: \mu_{BB} \ne \mu_{BA} \\
\end{equation}

Observed Samples
--

```python
library(ggplot2)
library(reshape2)
library(plyr)

# Blood pressure of the heterozygous (BA) genotype
x = c(86, 88, 89, 89, 92, 93, 94, 94, 94, 95, 95, 96, 96, 97, 97, 98, 98, 99, 99, 101, 106, 107, 110, 113, 116, 118)

# Blood pressure of the homozygous (BB) genotype
y = c(89, 90, 92, 93, 93, 96, 99, 99, 99, 102, 103, 104, 105, 106, 106, 107, 108, 108, 110, 110, 112, 114, 116, 116)

# Observed difference between the means of the two genotypes
diff_observed = mean(y) - mean(x)
print(diff_observed)
```

```python
blood_pressure = c(x, y)
genotype = c(rep("BA", length(x)), rep("BB", length(y)))
df = data.frame(genotype, blood_pressure)
head(df)
tail(df)
```

```python
p = ggplot(df)
p = p + geom_boxplot(aes(y = blood_pressure, x = genotype, fill = genotype), alpha = 0.75)
p = p + labs(fill = "Genotype", title = "Observed (actual) Samples")
p = p + xlab("Genotype")
p = p + ylab("Blood Pressure")
print (p)
```

Obviously, from the above boxplot, there is a difference between the two datasets. Using a **t-test**:

```python
t.test(x,y)
```

According to the $p{-}value$ of the t-test
\begin{align}
\because p{-}value < 0.05 \\
\therefore H_O\text{ is rejected} \\
\end{align}

Bootstrapping
--
Now, let's estimate the confidence interval and the $p{-}value$ using bootstrap resampling using 10,000 replicates (iterations).

```python
# Number of replicates
number_of_replicates = 10000

# Bootstrapped Heterozygous (BA)
x_bootstrap_mean = NULL

# Bootstrapped Homozygous (BB)
y_bootstrap_mean = NULL

# Bootstrapped Differences
diff_bootstrap = NULL

for (i in 1 : number_of_replicates) {
    # Sample with replacement for BA and BB
    x_bootstrap = sample  (x, length(x), TRUE)
    y_bootstrap = sample  (y, length(y), TRUE)

    x_bootstrap_mean[i] = mean(x_bootstrap)
    y_bootstrap_mean[i] = mean(y_bootstrap)

    # Difference between means of bootstrapped sets
    diff_bootstrap[i] = mean(y_bootstrap) - mean(x_bootstrap)
}

df = data.frame (genotype = c (rep("BA", number_of_replicates),
                                rep("BB", number_of_replicates)),
                blood_pressure = c (x_bootstrap_mean, y_bootstrap_mean))

head(df)
tail(df)
```

Means of "Bootstrapped" Samples
--

```python
cdf = ddply(df, "genotype", summarise, blood_pressure_mean = mean(blood_pressure))

p = ggplot(df, aes(blood_pressure, fill = genotype))
p = p + geom_density(alpha = 0.75)
p = p + geom_vline(data=cdf, aes(xintercept = blood_pressure_mean, colour = genotype), linetype="dashed")
p = p + theme(legend.position="top")
p = p + labs(fill = "Genotype", title = "Bootstrapped Samples")
p = p + xlab("Blood Pressure")
print(p)
```

Confidence Interval
--

```python
alpha = 0.05
ci = t(data.frame(quantile(diff_bootstrap, c(alpha/2, 1 - alpha/2))))
ci[1]
ci[2]
```

Difference between Means of Bootstrapped Samples
--

```python
df1 = data.frame (diff_bootstrap)

lines = c (diff_observed, mean(diff_bootstrap), ci[1], ci[2])
names = c ("Observed", "Bootstrapped", "CI Lower", "CI Upper")

df2 = data.frame(names, lines)

colors = c("#984ea3", "#ff7f00", "#e41a1c", "#377eb8")

p = ggplot()
p = p + geom_density (data = df1, aes(diff_bootstrap), fill = "#4daf4a", alpha = 0.75)
p = p + geom_vline   (data = df2, aes(xintercept=lines, colour=names), linetype=c("solid", "dashed", "dashed", "dashed"), size=c(1.5, 0.5, 0.5, 0.5))
p = p + scale_colour_manual(values = colors)
p = p + theme(legend.position="top")
p = p + labs(title = "Bootstrapped Differences")
p = p + xlab("Differences")
print(p)
```

Null Differences
--
Under the null hypothesis, there is no difference between the means of the two genotypes.

```python
# Difference between means under null hypothesis (i.e., zero)
diff_null = diff_bootstrap - mean(diff_bootstrap)

df = data.frame(diff_null)

critical.index = max(number_of_replicates * alpha / 2, 1)
critical.value.left = diff_null[order(diff_null)][critical.index]
critical.value.right = diff_null[order(diff_null, decreasing=TRUE)][critical.index]

p = ggplot()
p = p + geom_density(data = df, aes(diff_null), fill = "#4daf4a", alpha = 0.75)
p = p + geom_vline(xintercept=diff_observed, colour = "black", linetype="solid", size=0.5)
p = p + geom_vline(xintercept=critical.value.left, colour = "red", linetype="solid", size=0.5)
p = p + geom_vline(xintercept=critical.value.right, colour = "red", linetype="solid", size=0.5)
p = p + geom_vline(data=df, aes(xintercept=mean(diff_null)), colour="blue", linetype="dashed", size=0.5)
p = p + labs(title = "Null Differences")
p = p + xlab("Differences")
print(p)
```

*P*-value
--
*P*-value is the ratio of how many times the difference under the null hypothesis ≥ the observed difference to the total number iterations.

```python
pvalue = sum(abs(diff_null) >= abs(diff_observed)) / number_of_replicates
print (pvalue)
```

Now according to the bootstrap test,

\begin{equation}
\because p{-}value < 0.05 \\
\therefore H_O\text{ is rejected} \\
\end{equation}

Which is agreeing with the earlier t-test.
